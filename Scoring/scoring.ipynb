```python
import pandas as pd
```

```python
df = pd.read_csv("..\data\students_100.csv")
```

```python
# Normalization
def normalization(df, min, max):
    return ((df - min) / (max - min)) * 100

def inverse_normalization(df, min, max):
    return ((1 - ((df - min) / (max - min)))) * 100
```

## Normalization

```python
df['gpa_n'] = normalization(df['gpa'], 0, 10)
df['gpa_n']
```

```python
df['attendance_n'] = normalization(df['attendance'],0 , 100)
df['attendance_n']
```

```python
df['assignments_completion_n'] = normalization(df['assignments_completion'], 0, 100)
df['assignments_completion_n']
```

```python
df['stress_level_n'] = inverse_normalization(df['stress_level'],1,10)
df['stress_level_n']
```

```python
df['sleep_hours_n'] = normalization(df['sleep_hours'],0,10)
df['sleep_hours_n']
```

```python
df['mental_wellbeing_n'] = normalization(df['mental_wellbeing'],1,10)
df['mental_wellbeing_n']
```

```python
df['productivity_score_n'] = normalization(df['productivity_score'],1, 10)
df['productivity_score_n']
```

```python
df['distractions_n'] = inverse_normalization(df['distractions'], 1, 10)
df['distractions_n']
```

```python
df['career_clarity_n'] = normalization(df['career_clarity'], 1, 10)
df['career_clarity_n']
```

```python
df['skill_readiness_n'] = normalization(df['skill_readiness'], 1 , 10)
df['skill_readiness_n']
```

```python
df['engagement_score_n'] = normalization(df['engagement_score'], 0, 100)
df['engagement_score_n']
```

## Score Calculation

```python
df['APS'] = (
    0.45 * df['gpa_n'] +
    0.25 * df['attendance_n'] +
    0.20 * df['assignments_completion_n'] +
    0.10 * df['engagement_score_n']
)
df['APS']
```

```python
df['WWS'] = (
    0.40 * df['stress_level_n'] +
    0.30 * df['sleep_hours_n'] +
    0.30 * df['mental_wellbeing_n']
)
df['WWS']
```

```python
df['PTMS'] = (
    0.60 * df['productivity_score_n'] +
    0.40 * df['distractions_n']
)
df['PTMS']
```

```python
df['CRS'] = (
    0.50 * df['career_clarity_n'] +
    0.50 * df['skill_readiness_n']
)
df['CRS']
```

```python
df['SRI'] = (
    0.30 * df['APS'] +
    0.25 * df['WWS'] +
    0.20 * df['PTMS'] +
    0.25 * df['CRS']
)
df['SRI']
```

```python
def classify_sri(sri):
    if sri >= 80:
        return "Green"
    elif sri >= 65:
        return "Blue"
    elif sri >= 50:
        return "Yellow"
    else:
        return "Red"
```

```python
df['Risk_Category'] = df['SRI'].apply(classify_sri)
```

```python
# Save output
df.to_csv("../data/students_norm.csv", index = False)
```
